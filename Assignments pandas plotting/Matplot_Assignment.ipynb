```python
import pandas as pd
import matplotlib.pyplot as plt
import logging

logging.basicConfig(filename="graph_generation.log",level=logging.INFO,format="%(asctime)s - %(levelname)s - %(message)s")
logging.info("Script started: Loading dataset...")
try:
    df = pd.read_excel("Superstore.xlsx")
    df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
    df["order_date"] = pd.to_datetime(df["order_date"])
    logging.info("Dataset loaded successfully.")
except Exception as e:
    logging.error(f"Error loading dataset: {e}")

def generate_graphs(df):
    plt.style.use("ggplot")
    try:
        logging.info("Generating Sales by Category graph...")
        plt.figure(figsize=(8, 5))
        df.groupby("category")["sales"].sum().plot(kind="bar", color=['blue', 'green', 'red'])
        plt.title("Total Sales by Category")
        plt.ylabel("Total Sales")
        plt.xlabel("Category")
        plt.xticks(rotation=45)
        plt.savefig("sales_by_category.png")
        plt.show()
        logging.info("Sales by Category graph saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Sales by Category graph: {e}")
    try:
        logging.info("Generating Profit by Category graph...")
        plt.figure(figsize=(8, 5))
        df.groupby("category")["profit"].sum().plot(kind="bar", color=['purple', 'orange', 'cyan'])
        plt.title("Total Profit by Category")
        plt.ylabel("Total Profit")
        plt.xlabel("Category")
        plt.xticks(rotation=45)
        plt.savefig("profit_by_category.png")
        plt.show()
        logging.info("Profit by Category graph saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Profit by Category graph: {e}")
    try:
        logging.info("Generating Sales by Region graph...")
        plt.figure(figsize=(8, 5))
        df.groupby("region")["sales"].sum().plot(kind="bar", color="teal")
        plt.title("Total Sales by Region")
        plt.ylabel("Sales")
        plt.xlabel("Region")
        plt.savefig("sales_by_region.png")
        plt.show()
        logging.info("Sales by Region graph saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Sales by Region graph: {e}")
    try:
        logging.info("Generating Orders by Ship Mode Pie Chart...")
        plt.figure(figsize=(8, 5))
        df["ship_mode"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=['lightblue', 'pink', 'yellow', 'gray'])
        plt.title("Orders by Shipping Mode")
        plt.ylabel("")
        plt.savefig("orders_by_shipmode.png")
        plt.show()
        logging.info("Orders by Ship Mode Pie Chart saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Orders by Ship Mode Pie Chart: {e}")
    try:
        logging.info("Generating Sales vs Profit Scatter Plot...")
        plt.figure(figsize=(8, 5))
        plt.scatter(df["sales"], df["profit"], alpha=0.6, color="red", edgecolors="black")
        plt.title("Sales vs Profit")
        plt.xlabel("Sales")
        plt.ylabel("Profit")
        plt.savefig("sales_vs_profit.png")
        plt.show()
        logging.info("Sales vs Profit Scatter Plot saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Sales vs Profit Scatter Plot: {e}")
    try:
        logging.info("Generating Discount Distribution Histogram...")
        plt.figure(figsize=(8, 5))
        plt.hist(df["discount"], bins=20, color="blue", edgecolor="black")
        plt.title("Discount Distribution")
        plt.xlabel("Discount")
        plt.ylabel("Frequency")
        plt.savefig("discount_distribution.png")
        plt.show()
        logging.info("Discount Distribution Histogram saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Discount Distribution Histogram: {e}")
    try:
        logging.info("Generating Sales by Sub-Category graph...")
        plt.figure(figsize=(12, 5))
        df.groupby("sub-category")["sales"].sum().sort_values().plot(kind="barh", color="orange")
        plt.title("Total Sales by Sub-Category")
        plt.xlabel("Sales")
        plt.ylabel("Sub-Category")
        plt.savefig("sales_by_subcategory.png")
        plt.show()
        logging.info("Sales by Sub-Category graph saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Sales by Sub-Category graph: {e}")
    try:
        logging.info("Generating Monthly Sales Trend Line Chart...")
        df.set_index("order_date", inplace=True)
        plt.figure(figsize=(10, 5))
        df["sales"].resample("M").sum().plot(marker="o", linestyle="-", color="red")
        plt.title("Monthly Sales Trend")
        plt.xlabel("Date")
        plt.ylabel("Sales")
        plt.savefig("sales_trend.png")
        plt.show()
        logging.info("Monthly Sales Trend Line Chart saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Monthly Sales Trend Line Chart: {e}")
    try:
        logging.info("Generating Top 10 Cities by Sales graph...")
        plt.figure(figsize=(10, 5))
        df.groupby("city")["sales"].sum().nlargest(10).plot(kind="bar", color="green")
        plt.title("Top 10 Cities by Sales")
        plt.xlabel("City")
        plt.ylabel("Sales")
        plt.xticks(rotation=45)
        plt.savefig("top_cities_sales.png")
        plt.show()
        logging.info("Top 10 Cities by Sales graph saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Top 10 Cities by Sales graph: {e}")
    try:
        logging.info("Generating Quantity Ordered Distribution Histogram...")
        plt.figure(figsize=(8, 5))
        plt.hist(df["quantity"], bins=10, color="purple", edgecolor="black")
        plt.title("Quantity Ordered Distribution")
        plt.xlabel("Quantity")
        plt.ylabel("Frequency")
        plt.savefig("quantity_distribution.png")
        plt.show()
        logging.info("Quantity Ordered Distribution Histogram saved successfully.")
    except Exception as e:
        logging.error(f"Error generating Quantity Ordered Distribution Histogram: {e}")

    logging.info("All graphs generated successfully.")

generate_graphs(df)
```
