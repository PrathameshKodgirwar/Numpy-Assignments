# 14. Featching records from June 2014

```python
print("Records for June 2014:")
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
june_data = data[data['Order_Date'].dt.strftime('%Y-%m') == '2014-06']
print(june_data)
```

# 15. Calculate month-wise average profit


```python
print("Month-wise average profit:")
data['Month'] = data['Order_Date'].dt.to_period('M')
avg_profit = data.groupby('Month')['Profit'].mean()
print(avg_profit)
```

# 26. Create a DataFrame where order date and ship date both are in JAN-2014


```python
import pandas as pd
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
data['Ship_Date'] = pd.to_datetime(data['Ship_Date'])
janData = data[(data['Order_Date'].dt.strftime('%Y-%m') == '2014-01') &  (data['Ship_Date'].dt.strftime('%Y-%m') == '2014-01')]
print(janData)
```

# 27. Find records having profit for consumer segment




```python
consumerProfit = data[(data['Segment']=='Consumer')&(data['Profit']>0)] 
print(consumerProfit)
```

# 28. Generate a new DataFrame for every state (dynamic variable creation)




```python
stateDataFrame = {state:data[data['State']==state] for state in data['State'].unique()}
print(stateDataFrame)
```

# 29. Generate the region-wise, state-wise average profit and sales, and extract information for the East region

```python
regionAvg = data.groupby(['Region','State'])[['Profit','Sales']].mean().reset_index()
eastData = regionAvg[regionAvg['Region']=='East']
print(eastData)
```

# 30. Find all records for furniture delivered in Texas


```python
furnitureTexas=data[(data['State']=='Texas')&(data['Category']=='Furniture')]
print("DataFrame where Order Date and Ship Date are in JAN-2014:")
print(janData)
print("\nRecords with profit for Consumer segment:")
print(consumerProfit)
print("\nState-wise DataFrames:")
for state, df in stateDataFrame.items():
    print(f"State:{state}")
    print(df.head())
print("\nRegion-wise,State-wise Average Profit and Sales(East Region):")
print(eastData)
print("\nFurniture records delivered in Texas:")
print(furnitureTexas)
```
